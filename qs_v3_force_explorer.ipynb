# V3 Kite Quasi-Steady Force Explorer
Play with the sliders to recompute the quasi-steady solution and visualise the force balance for the V3 kite.

```python
import sys
from pathlib import Path
import matplotlib.pyplot as plt
import ipywidgets as widgets
from qs_state_v3_forces import plot_qs_forces
from IPython.display import display

HERE = Path.cwd()  # notebook’s directory
if str(HERE) not in sys.path:
    sys.path.insert(0, str(HERE))



    
# Create widgets with proper spacing for value display
style = {'description_width': '50px'}
slider_layout = widgets.Layout(width='300px')

course_slider = widgets.FloatSlider(min=0, max=360, step=5, value=90, description='Course °:', style=style, layout=slider_layout, readout=True, readout_format='.1f')
mass_slider = widgets.FloatSlider(min=0, max=120, step=5, value=60, description='Mass kg:', style=style, layout=slider_layout, readout=True, readout_format='.1f')
elevation_slider = widgets.FloatSlider(min=-30, max=60, step=2, value=0, description='Elev. °:', style=style, layout=slider_layout, readout=True, readout_format='.1f')
azimuth_slider = widgets.FloatSlider(min=-90, max=90, step=2, value=0, description='Azim. °:', style=style, layout=slider_layout, readout=True, readout_format='.1f')
wind_slider = widgets.FloatSlider(min=4, max=25, step=1, value=12, description='Wind m/s:', style=style, layout=slider_layout, readout=True, readout_format='.1f')
radial_slider = widgets.FloatSlider(min=-10, max=6, step=0.5, value=0, description='Radial m/s:', style=style, layout=slider_layout, readout=True, readout_format='.1f')

# Give the controls VBox enough width
controls_layout = widgets.Layout(width='300px', min_width='300px')
controls = widgets.VBox([course_slider, mass_slider, elevation_slider, azimuth_slider, wind_slider, radial_slider], layout=controls_layout)
out = widgets.Output()

def render_forces(change=None):
    with out:
        out.clear_output(wait=True)
        fig, axes, state, forces = plot_qs_forces(
            course_deg=course_slider.value,
            mass_wing=mass_slider.value,
            elevation_deg=elevation_slider.value,
            azimuth_deg=azimuth_slider.value,
            wind_speed=wind_slider.value,
            speed_radial=radial_slider.value,
            show=False,
            close=False,
        )
        display(fig)
        plt.close(fig)

for slider in [course_slider, mass_slider, elevation_slider, azimuth_slider, wind_slider, radial_slider]:
    slider.observe(render_forces, names='value')

render_forces()
display(widgets.HBox([controls, out]))
```
